```python
danmaku_filter = [
    "赞",
    "妙啊",
    "有点东西",
    "很有精神",
    "泪目",
    "赢麻了",
    "钝角",
    "干杯",
    "1",
    "来了来了",
    "原地去世",
    "永远爱你",
    "嗷呜",
    "第一时间直播最新最热单机游戏，欢迎关注",
    "直播通知群（禁言）：104294141",
    "爱你",
    "吃零食",
    "打电动",
    "晚安喵",
    "棒",
    "(｀・ω・´)",
    "打卡",
]
danmaku_equivalence = {
    "？": [
        "？？",
        "？？？",
        "？？？？",
        "？？？？？",
        "？？？？？？",
        "?",
        "??",
        "???",
        "????",
        "?????",
        "??????",
    ],
    "乐": [
        "哈哈",
        "哈哈哈",
        "哈哈哈哈",
        "哈哈哈哈哈",
        "哈哈哈哈哈哈",
        "哈哈哈哈哈哈哈",
        "哈哈哈哈哈哈哈哈",
        "哈哈哈哈哈哈哈哈哈",
        "哈哈哈哈哈哈哈哈哈哈",
        "233",
        "2333",
        "23333",
        "233333",
        "2333333",
    ],
    "草": ["艹", "cao"],
    "甜": ["甜甜甜", "kdl", "磕到了", "磕磕磕", "ttk", "偷偷磕"],
    "好": ["好好好"],
}
```

```python
p = {}
time_list_tot = []
for d in danmaku:
    text = d["text"]
    time = round(float(d["time"]))
    # filter danmaku
    if text in danmaku_filter:
        continue
    # combine equivalent danmaku
    for key in danmaku_equivalence:
        if text in danmaku_equivalence[key]:
            text = key
            break
    # add danmaku to pool
    if text in p:
        p[text].append(time)
    else:
        p[text] = [time]
    # add time to total time list
    time_list_tot.append(time)

time_list_text = p
time_list_text_sorted = [{"text": text, "time": p[text]} for text in p]


def sort_key(e: Dict[str, any]):
    return len(e["time"])


time_list_text_sorted.sort(reverse=True, key=sort_key)

for item in time_list_text_sorted[:100]:
    print(item["text"], len(item["time"]))

```

```python
def calc_freq_map(time_list: List[int], max_time: int):
    average_duration = 10
    map_step_size = 5

    # the input list should have been already in ascending order, but we make sure about it again
    time_list.sort()
    # calculate the time averaged frequency
    x = [_ for _ in range(0, max_time, map_step_size)]
    y = [0 for _ in range(0, max_time, map_step_size)]
    hi, lo = 0, 0
    max_index = len(time_list) - 1
    for i in range(len(x)):
        left = x[i] - average_duration / 2
        right = x[i] + average_duration / 2
        while time_list[lo] < left and lo < max_index:
            lo += 1
        while time_list[hi] < right and hi < max_index:
            hi += 1
        y[i] = (hi - lo) / average_duration

    return x, y
```

```python
# selected_danmaku = ["乐", "？"]
selected_danmaku = [item["text"] for item in time_list_text_sorted[:20]]

time_list_selected = {"All": time_list_tot}
for text in selected_danmaku:
    time_list_selected[text] = time_list_text[text]

data = []
max_time = max(time_list_tot)
average_frequency = len(time_list_tot) / max_time
data.append(["3xAverage", 0, 3 * average_frequency])
data.append(["3xAverage", max_time, 3 * average_frequency])
for text in time_list_selected:
    x, y = calc_freq_map(time_list_selected[text], max_time)
    for i in range(len(x)):
        data.append([text, x[i], y[i]])
df = pd.DataFrame(data, columns=["Danmaku", "Time", "Frequency"])

fig = px.line(df, x="Time", y="Frequency", color="Danmaku")
fig.show()
```
