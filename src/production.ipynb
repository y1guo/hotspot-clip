# Detect Hotspots

```python
hotspots = {}
for pool in danmaku_pool.segments():
    for roomid in pool.df["roomid"].unique():
        start_time = TIMEZONE.localize(datetime.datetime.fromtimestamp(pool.df["time"].min())).strftime("%m/%d %H:%M:%S")
        end_time = TIMEZONE.localize(datetime.datetime.fromtimestamp(pool.df["time"].max())).strftime("%m/%d %H:%M:%S")
        print("roomid: {}, {}-{}".format(roomid, start_time, end_time))
        time, density = pool.get_danmaku_density(roomid, KERNEL_SIGMA)
        _hotspots = get_hotspots(time, density, KERNEL_SIGMA)
        if roomid in hotspots:
            hotspots[roomid] = pd.concat([hotspots[roomid], _hotspots])
        else:
            hotspots[roomid] = _hotspots
        print("number of hotspots:", len(_hotspots))
        
```

```python
# save hotspots to with pickle
import pickle
with open("cache/hotspots.pickle", "wb") as f:
    pickle.dump(hotspots, f)
```

```python
# load hotspots from pickle
import pickle
with open("cache/hotspots.pickle", "rb") as f:
    hotspots = pickle.load(f)
```
